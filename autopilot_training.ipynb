### IMPORTS

```python
import torch
import torchvision
from tqdm import tqdm
from sklearn.metrics import mean_squared_error
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

from autopilot_dataset import AutopilotDataset
from autopilot_model import AutopilotModel
```

```python
import os
os.environ["CUDA_DEVICE_ORDER"]="PCI_BUS_ID"
os.environ["CUDA_VISIBLE_DEVICES"]="0"
```

<br>

### HYPERPARAMETERS

```python
BATCH_SIZE = 256
MAX_EPOCHS = 100
EARLY_STOPPING_PATIENCE = 25

MODELS_DIR = "/home/greg/models/"
DATASETS_DIR = "/home/greg/datasets/"
NAME = "camera-basic_motion_0"
MODEL_PATH = MODELS_DIR + NAME + "_resnet18.pth"
TRAINING_DATASET = DATASETS_DIR + NAME + "_training/"
VALIDATION_DATASET = DATASETS_DIR + NAME + "_validation/"
TESTING_DATASET = DATASETS_DIR + NAME + "_testing/"

TRANSFORMS = torchvision.transforms.Compose([
    torchvision.transforms.ColorJitter(0.25, 0.25, 0.25, 0.25),
    torchvision.transforms.ToTensor(),
    torchvision.transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])
```

<br>

### DATA

```python
training_dataset = AutopilotDataset(TRAINING_DATASET, TRANSFORMS, random_horizontal_flip=True, random_noise=True, random_blur=True)
training_loader = torch.utils.data.DataLoader(training_dataset, batch_size=BATCH_SIZE, shuffle=True)

validation_dataset = AutopilotDataset(VALIDATION_DATASET, TRANSFORMS, random_horizontal_flip=False, random_noise=False, random_blur=False)
validation_loader = torch.utils.data.DataLoader(validation_dataset, batch_size=BATCH_SIZE, shuffle=False)

testing_dataset = AutopilotDataset(TESTING_DATASET, TRANSFORMS, random_horizontal_flip=False, random_noise=False, random_blur=False)
testing_loader = torch.utils.data.DataLoader(testing_dataset, batch_size=1, shuffle=False)
```

<br>

### MODEL

```python
model = AutopilotModel()
optimizer = torch.optim.Adam(model.parameters())
loss_function = torch.nn.MSELoss()
```

<br>

### TRAINING

```python
training_losses = []
validation_losses = []
epochs_without_improvement = 0

def plot_losses():
    fig, ax = plt.subplots()
    ax.plot([x for x in range(len(training_losses))], training_losses, label='training_loss')
    ax.plot([x for x in range(len(validation_losses))], validation_losses, label='validation_loss')
    ax.set(xlabel='epochs', ylabel='loss', title='Training Progress')
    ax.grid()
    plt.legend()
    plt.show()
    
def run_epoch(tepoch, name, training):
    loss = 0.0
    iterations = 0
    
    for images, annotations in tepoch:
        tepoch.set_description(f"{name} Epoch {epoch}")

        images = images.cuda()
        annotations = annotations.cuda()
        
        if training:
            optimizer.zero_grad()
            model.train()
            outputs = model(images)
        else:
            with torch.no_grad():
                model.eval()
                outputs = model(images)
        
        loss = loss_function(outputs, annotations)
        
        if training:
            loss.backward()
            optimizer.step()

        loss += loss.item()
        iterations += 1

    return float((loss/iterations).cpu().detach().numpy())

for epoch in range(MAX_EPOCHS):
    with tqdm(training_loader, unit="batch") as training_epoch:
        avg_training_loss = run_epoch(training_epoch, "Training", training=True)
        training_losses.append(avg_training_loss)
    
    with tqdm(validation_loader, unit="batch") as validation_epoch:
        avg_validation_loss = run_epoch(validation_epoch, "Validation", training=False)
        validation_losses.append(avg_validation_loss)
        
        if avg_validation_loss <= np.min(validation_losses):
            epochs_without_improvement = 0
            print("validation loss decreased to " + str(avg_validation_loss) + ", saving model")
            model.save_to_path(MODEL_PATH)
        else:
            epochs_without_improvement += 1
            if epochs_without_improvement >= EARLY_STOPPING_PATIENCE:
                print("validation loss hasn't improved in last " + str(EARLY_STOPPING_PATIENCE) + " epochs, stopping training")
                break
        
    
    plot_losses()
```

<br>

### TESTING

```python
steering_true = []
steering_predicted = []

throttle_true = []
throttle_predicted = []

with torch.no_grad():
    model.eval()
    for images, annotations in testing_loader:
        images = images.cuda()

        outputs = model(images)

        for y_true in annotations.cpu().numpy():
            steering_true.append(y_true[0])
            throttle_true.append(y_true[1])

        for y_predicted in outputs.cpu().numpy():
            steering_predicted.append(y_predicted[0])
            throttle_predicted.append(y_predicted[1])

for output in [(steering_true, steering_predicted, "steering"), (throttle_true, throttle_predicted, "throttle")]:
    true, predicted, name = output
    mse = mean_squared_error(true, predicted)
    print(name + " mse :", mse)
```
